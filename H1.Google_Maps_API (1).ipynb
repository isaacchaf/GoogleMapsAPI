A. Use the [*Nearby Seach*](https://developers.google.com/places/web-service/search) Method to find all the restaurant near to the UPY (20.988459, -89.736768) around 2km. In order to fulfill this item you need to:
* Create a *search_nearby_places* function with the appropiate parameters. It must return a JSON with the results. Places must be ranked by distance.
* Create a table with the following columns: Name, Place Id, Rating, Place Types, Total of User Ratings.

```python
#  Import the required libaries 
import requests
import json
import pandas as pd
import urllib.parse
```

```python
def search_nearby_place(latitude, longitude, place, key):
    """Find the places nearby a location .

    Args:
      latitude: An float value for the latitude.
      longitude: An float value for the longitude.
      place: A string for the type of place is looked for.
      key: A string of the Google API Key.
   
    Returns:
      The results of the search as a json.
    """
    url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={},{}&radius=2000&type={}&key={}'.format(
        latitude, longitude, place, key)
    res = requests.get(url)
    results = json.loads(res.content)
    return results     
```

```python
google_api_key = 'Your API key'
raw_data = search_nearby_place(20.988459, -89.736768, 'restaurant', google_api_key)
```

```python
restaurantes_yucatan = pd.read_json(json.dumps(raw_data['results']), orient = 'records')
restaurantes_table = restaurantes_yucatan[['name', 'place_id', 'rating', 'types','user_ratings_total']]
restaurantes_table
```

B. Create a function to Find the Place Id given Place Name (String) and Location (Lat/Lont Tuple).

```python
def find_place_id(placeName, location, key):
    """Find a place id given the name and the location.

    Args:
      placeName: A string for the name of a place.
      location: A tuple with two float values corresponding to the latitude and longitude.
      key: A string of the Google API Key.
   
    Returns:
      The results of the search as a json.
    """
    placeName = urllib.parse.quote(placeName) # Eliminate the spaces of the name string
    url = 'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={}&inputtype=textquery&locationbias=circle:600@{},{}&key={}'.format(
        placeName, location[0], location[1], key)
    res = requests.get(url)
    results = json.loads(res.content)
    return results     
```

D. Use the [*Place Details*](https://developers.google.com/places/web-service/details#PlaceDetailsResults) Method to find the Reviews of a Place in English and Spanish. In order to fulfill this item you need to:
* Create a *find_reviews* function with the appropiate parameters. It must return a JSON with the results. 
* Create a table with the following columns: Author Name, Language, Rating, Text, Time.

```python
def find_reviews(place_id, language, key):
    """Find the reviews of a given place.

    Args:
      place_id: A string for the place id.
      language: A string with the abreviation of the lenguage of the reviews.
      key: A string of the Google API Key.
   
    Returns:
      The results of the search as a json.
    """
    url = 'https://maps.googleapis.com/maps/api/place/details/json?place_id={}&language={}&fields=reviews&key={}'.format(
        place_id, language, key)
    res = requests.get(url)
    results = json.loads(res.content)
    return results     
```

```python
google_api_key = 'Your API key'

reviews_Stabucks_es = find_reviews('ChIJaUUfylFxVo8RgEGjoansPbc', 'es', google_api_key)
reviews_Stabucks_en = find_reviews('ChIJaUUfylFxVo8RgEGjoansPbc', 'en', google_api_key)
```

```python
starbucks_reviews_es = pd.read_json(json.dumps(reviews_Stabucks_es['result']['reviews']), orient = 'records')
starbucks_reviews_es = starbucks_reviews_es[['author_name', 'language', 'rating', 'text','time']]

starbucks_reviews_en = pd.read_json(json.dumps(reviews_Stabucks_en['result']['reviews']), orient = 'records')
starbucks_reviews_en = starbucks_reviews_en[['author_name', 'language', 'rating', 'text','time']]

starbucks_reviews = pd.concat([starbucks_reviews_es, starbucks_reviews_en])
starbucks_reviews
```

E. Find similar places to Los Trompos Circuito. Rank them by its "Prominence". In order to fulfill this item you need to:
* Create a *find_similar* function with the appropiate parameters. It must return a JSON with the results.
* Create a table with the following columns: Place Name, Plaece Id, Rating, Total of Use Ratings.

```python
def find_similar(name, location, key):
    """Find similar places give the name of a place.

    Args:
      name: A string for the name of a place.
      location: A tuple with two float values corresponding to the latitude and longitude.
      key: A string of the Google API Key.
   
    Returns:
      The results of the search as a json.
    """    

    url = 'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={}&inputtype=textquery&fields=types&key={}'.format(
    name, key)
    res = requests.get(url)
    results = json.loads(res.content)
    place_type = results['candidates'][0]['types'][0]
    
    url2 ='https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={},{}&radius=2000&rankby=prominence&type={}&key={}'.format(
        location[0], location[1], place_type, key)
    res2 = requests.get(url2)
    results_nearby = json.loads(res2.content)
    return results_nearby 
```

```python
google_api_key = 'Your API key'
similar_data = find_similar('Los Trompos Circuito', (20.988459, -89.736768), google_api_key)
```

```python
similar_place = pd.read_json(json.dumps(similar_data['results']), orient = 'records')
similar_place_table = similar_place[['name', 'place_id', 'rating', 'user_ratings_total']]
similar_place_table
```
